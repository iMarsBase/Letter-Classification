```python
import numpy as np
import pandas as pd
f2 = open("letter-recognition.text","r")
lines = f2.readlines()
lines[1]
```

# New Section

```python
let_tr = []
x_box = []
y_box = []
width = []
high = []
on_pix = []
x_bar = []
y_bar = []
x2_bar = []
y2_bar = []
xy_bar = []
x2y_bar = []
xy2_bar = []
x_ege = []
xeg_vy = []
y_ege = []
yeg_vx = []
```

```python
def split_data(data):
    df_data = pd.DataFrame()
    for i in range(len(data)):
        sample = data[i].split(",")
        let_tr.append(str(sample[0]))
        x_box.append(int(sample[1]))
        y_box.append(int(sample[2]))
        width.append(int(sample[3]))
        high.append(int(sample[4]))
        on_pix.append(int(sample[5]))
        x_bar.append(int(sample[6]))
        y_bar.append(int(sample[7]))
        x2_bar.append(int(sample[8]))
        y2_bar.append(int(sample[9]))
        xy_bar.append(int(sample[10]))
        x2y_bar.append(int(sample[11]))
        xy2_bar.append(int(sample[12]))
        x_ege.append(int(sample[13]))
        xeg_vy.append(int(sample[14]))
        y_ege.append(int(sample[15]))
        yeg_vx.append(int(sample[16][:-1]))
    df_data['let_tr'] = let_tr
    df_data['x_box'] = x_box
    df_data['y_box'] = y_box
    df_data['width'] = width
    df_data['high'] = high
    df_data['on_pix'] = on_pix
    df_data['x_bar'] = x_bar
    df_data['y_bar'] = y_bar
    df_data['x2_bar'] = x2_bar
    df_data['y2_bar'] = y2_bar
    df_data['xy_bar'] = xy_bar
    df_data['x2y_bar'] = x2y_bar
    df_data['xy2_bar'] = xy2_bar
    df_data['x_ege'] = x_ege
    df_data['xeg_vy'] = xeg_vy
    df_data['y_ege'] = y_ege
    df_data['yeg_vx'] = yeg_vx

    return df_data
```

```python
df = split_data(lines)
df
```
